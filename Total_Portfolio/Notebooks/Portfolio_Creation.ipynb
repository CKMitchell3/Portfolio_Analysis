# **Total Portfolio Analysis**

## Import Libraries and Dependencies

```python
#Import your libraries and dependencies
import os
import pandas as pd
import yfinance as yf
import numpy as np
import pandas_datareader as pdr

%matplotlib inline
```

## Import Yahoo Stock Data

```python
#Set your Datareader
yf.pdr_override()

#Set your tickers
tickers = ["SQ", "TSLA", "AAPL", "FB", "PYPL", "SHOP", "TDOC", "NVDA", "TWTR", "Z"]

#Set your portfolio dates
start = '2016-01-01'
end = '2021-05-06'
```

```python
#Load your portfolio Dataframe into a variable
df = pdr.data.get_data_yahoo(tickers, start, end)
```

```python
df.head()
```

```python
df.info()
```

```python
total_portfolio = df[['Close']]

total_portfolio.head(10)
```

```python
total_portfolio.columns = ["AAPL", "FB", "NVDA", "PYPL", "SHOP", "SQ", "TDOC", "TSLA", "TWTR", "Z"]

total_portfolio.head(10)
```

```python
total_portfolio.count()
```

## Portfolio Volume Analysis

```python
portfolio_volumes = df[['Volume']]

portfolio_volumes.head(10)
```

```python
portfolio_volumes.columns = ["AAPL", "FB", "NVDA", "PYPL", "SHOP", "SQ", "TDOC", "TSLA", "TWTR", "Z"]

portfolio_volumes.tail(10)
```

### Year-To-Date Volume Analysis

```python
ytd_portfolio_volumes = portfolio_volumes.loc['2021']

ytd_portfolio_volumes.head(10)
```
