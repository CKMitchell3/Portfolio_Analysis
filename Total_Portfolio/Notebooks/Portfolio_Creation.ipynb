# **Portfolio Creation**

## Import Libraries and Dependencies

```python
#Import your libraries and dependencies
import os
import pandas as pd
import yfinance as yf
import numpy as np
import pandas_datareader as pdr

%matplotlib inline
```

## Import Yahoo Stock Portfolio Data

```python
#Set your Datareader
yf.pdr_override()

#Set your tickers
tickers = ["SQ", "TSLA", "AAPL", "FB", "PYPL", "SHOP", "TDOC", "NVDA", "TWTR", "Z"]

#Set your portfolio dates
start = '2016-01-01'
end = '2021-05-25'
```

```python
#Load your portfolio Dataframe into a variable
df = pdr.data.get_data_yahoo(tickers, start, end)
```

```python
df.tail()
```

```python
df.info()
```

```python
# Separate the Adjusted Closing Prices

stock_portfolio = df[['Adj Close']]

stock_portfolio.head(10)
```

```python
# Rename your columns

stock_portfolio.columns = ["AAPL", "FB", "NVDA", "PYPL", "SHOP", "SQ", "TDOC", "TSLA", "TWTR", "Z"]

stock_portfolio.head(10)
```

```python
stock_portfolio.count()
```

## Portfolio Volume Analysis

```python
stock_portfolio_volumes = df[['Volume']]

stock_portfolio_volumes.head(10)
```

```python
stock_portfolio_volumes.columns = ["AAPL", "FB", "NVDA", "PYPL", "SHOP", "SQ", "TDOC", "TSLA", "TWTR", "Z"]

stock_portfolio_volumes.tail(10)
```

### Year-To-Date Volume Analysis

```python
ytd_stock_portfolio_volumes = stock_portfolio_volumes.loc['2021']

ytd_stock_portfolio_volumes.head(10)
```

```python
ytd_stock_portfolio_volumes.tail()
```

```python
ytd_stock_portfolio_volumes.plot(figsize=(15,10))
```
