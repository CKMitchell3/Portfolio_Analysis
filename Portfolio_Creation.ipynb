```python
import os
import pandas as pd
import alpaca_trade_api as tradeapi
import numpy as np
import numpy as np
from dotenv import load_dotenv

%matplotlib inline
```

```python
load_dotenv("alpaca.env")
```

```python
#Amount of shares are calculated based on the closing price of each stock on April 23, 2018
#Each stock has $30,000 worth

#SQ closing price = $50.35
#TSLA closing price = $56.67
#AAPL closing price = $39.86
#GOOG closing price = $1,067.45
#DIS closing price = $97.40
#JPM closing price = $101.19
#TDOC closing price = $42.55

sq_shares = 596
tsla_shares = 529
aapl_shares = 753
goog_shares = 28
dis_shares = 308
jpm_shares = 297
tdoc_shares = 706
```

```python
#Create variables for each alpaca key to easily call it
alpaca_api_key = os.getenv("alpaca_api_key")
alpaca_secret_key = os.getenv("alpaca_secret_key")
```

```python
#Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
#Set your portfolio study dates
start = pd.Timestamp("2018-04-23", tz="America/New_York").isoformat()
end = pd.Timestamp("2021-05-05", tz="America/New_York").isoformat()

#Set your timeframe for the data
timeframe = "1D"

#Set your tickers
tickers = ["SQ", "TSLA", "AAPL", "GOOG", "DIS", "JPM", "TDOC"]
```

```python
my_portfolio = alpaca.get_barset(
    tickers,
    timeframe,
    limit = 1000,
    start = start,
    end = end
).df

my_portfolio.head(10)
```

```python
my_portfolio.count()
```

```python
SQ_portfolio = my_portfolio['SQ'][['close']]
```

```python
TSLA_portfolio = my_portfolio['TSLA'][['close']]
```

```python
AAPL_portfolio = my_portfolio['AAPL'][['close']]
```

```python
GOOG_portfolio = my_portfolio['GOOG'][['close']]
```

```python
DIS_portfolio = my_portfolio['DIS'][['close']]
```

```python
JPM_portfolio = my_portfolio['JPM'][['close']]
```

```python
TDOC_portfolio = my_portfolio['TDOC'][['close']]
```

```python
total_portfolio = pd.concat([SQ_portfolio, TSLA_portfolio, AAPL_portfolio, GOOG_portfolio, DIS_portfolio, JPM_portfolio, TDOC_portfolio], axis = 'columns', join = 'inner')

total_portfolio.head(10)
```
